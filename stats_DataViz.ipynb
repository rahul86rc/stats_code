```python
#Library Imports
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ds = sns.load_dataset('flights')
ds.head()
flights_may = ds.query('month == "May"')
flights_may
```

# LINE CHART TO SHOW TREND IN THE MONTH OF MAY

```python
sns.lineplot(data=ds,x='year',y='passengers', ci=None)
```

# PIE CHART TO GET PERCENTAGES OF MONTHWISE TRAVEL FOR 1960

```python
ds['year'].max()
pass_count = []

for month,df in ds[ds['year']==1960].groupby('month'):
        pass_count.append(df['passengers'].values[0])
        print (df['passengers'].values)
print (pass_count)
max_idx = pass_count.index(max(pass_count))
print (max_idx)
explode_dat = [0]*len(pass_count)
explode_dat[max_idx] = 0.2
month_dat = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
#Plot Pie Chart
plt.pie(pass_count, labels=month_dat, explode = explode_dat, autopct='%1.1f%%')
plt.show()
#Plot Donut Pie Chart
plt.pie(pass_count, labels=month_dat, explode = explode_dat, autopct='%1.1f%%')
centre_circle = plt.Circle((0, 0), 0.4, fc='white')
fig = plt.gcf()
# Adding Circle in Pie chart
fig.gca().add_artist(centre_circle)
plt.show()
```
